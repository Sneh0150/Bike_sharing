```python
#import libraries for visualization purpose

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# mapping season  for better visibilty
# season (1:spring, 2:summer, 3:fall, 4:winter)
bike['season'] = bike['season'].map({1:'spring',2:'summer',3:'fall',4:'winter'})

# mapping Year varible  for better visibilty
bike['yr'] = bike['yr'].map({0:'2018',1:'2019'})

# mapping Month varible for better visibilty
bike['mnth'] = bike['mnth'].map({1:'Jan',2:'Feb',3:'Mar',4:'April',
                                       5:'May',6:'June',7:'July',8:'Aug',
                                       9:'Sept',10:'Oct',11:'Nov',12:'Dec'})

```

```python
bike.head()
```

```python
# mapping weathersit for better clearity
#bike['weathersit']=bike.weathersit.map({1:'Clear',2:'Mist + Cloudy',3:'Light Snow',4:'Heavy Rain + Ice Pallets})
bike['weathersit']=bike.weathersit.map({1:"Good/Clear",2:'Moderate/Misty',3:'Bad/LightRain',4:'Worse/HeavyRain'}) 
# mapping weekday for better clearity
bike['weekday']=bike.weekday.map({0:'Sun',1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat'})
```

```python
# mapping workingday for better clearity

bike['workingday'] = bike['workingday'].map({0:'No',1:'Yes'})

```

```python
# mapping Holiday varible for better clearity

bike['holiday'] = bike['holiday'].map({0:'No',1:'Yes'})
```

### Model Building

```python
# Dividing training set into X_train and y_train sets for the model building

X_train = df_bike_train.drop(["cnt"],axis=1)
y_train = df_bike_train["cnt"]

print(y_train.shape)
print(X_train.shape)
```

```python
X_train.head()
```

```python
#importing RFE library for feature selection and after this will perform manual feature selection.

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# using RFE for feature selection and  limiting to selecton to 15 features

estimator = LinearRegression()
estimator.fit(X_train, y_train)

rfe = RFE(estimator,n_features_to_select=15)             
rfe = rfe.fit(X_train, y_train)
```

```python
#creating a list of features selected by RFE

list(zip(X_train, rfe.support_, rfe.ranking_))
```

```python
#creating a list of features selected by RFE

rfe_col = X_train.columns[rfe.support_]
rfe_col
```

```python
# Creating new train dataframe with RFE selected features

X_train_rfe = X_train[rfe_col]
X_train_rfe.head()
```
